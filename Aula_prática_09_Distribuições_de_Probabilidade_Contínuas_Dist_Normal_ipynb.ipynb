# Definição dos parâmetros da distribuição

```python
media = 30
desvio_padrao = 3

# Criando a distribuição Normal
dist = norm(loc=media, scale=desvio_padrao)
```

```python
# Valores para gráfico
valores = list(range(10, 50))

# Função densidade de probabilidade (PDF)
probabilities = [dist.pdf(valor) for valor in valores]
plt.plot(valores, probabilities, label='PDF')
plt.title('Função Densidade de Probabilidade - Normal')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Função distribuição acumulada (CDF)
acc = [dist.cdf(valor) for valor in valores]
plt.plot(valores, acc, label='CDF', color='orange')
plt.title('Função Distribuição Acumulada - Normal')
plt.xlabel('Valor')
plt.ylabel('Probabilidade Acumulada')
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Exemplo: Peso dos alunos normal com média 70kg e desvio padrão 5kg
peso_media = 70
peso_dp = 5
```

```python
# Calculando a probabilidade de peso estar entre 60 e 80 kg usando CDF
prob_peso = norm.cdf(x=80, loc=peso_media, scale=peso_dp) - norm.cdf(x=60, loc=peso_media, scale=peso_dp)
print(f'Probabilidade peso entre 60 e 80 kg: {prob_peso:.4f}')
```

```python
# Cálculo manual pelo z-score normal padrão
limite_inf = (60 - peso_media) / peso_dp
limite_sup = (80 - peso_media) / peso_dp
prob_peso_manual = norm.cdf(limite_sup) - norm.cdf(limite_inf)
print(f'Probabilidade manual pelo z-score: {prob_peso_manual:.4f}')
```
