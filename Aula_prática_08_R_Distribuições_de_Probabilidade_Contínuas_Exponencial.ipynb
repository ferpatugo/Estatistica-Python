```python
# Aula prática 08 - Python - Distribuições de Probabilidade Contínuas - Exponencial
from scipy.stats import expon
from matplotlib import pyplot as plt
import math
```

# Calculando função densidade de probabilidade (PDF)

```python
# Parâmetros da distribuição
beta = 30  # parâmetro de escala da exponencial

# Criando a distribuição exponencial
dist = expon(scale=beta)

# Valores para plotagem
valores = list(range(30, 50))

# Calculando função densidade de probabilidade (PDF)
probabilidades = [dist.pdf(valor) for valor in valores]
plt.plot(valores, probabilidades, label='PDF')
plt.title('Função Densidade de Probabilidade - Exponencial')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

# Calculando função distribuição acumulada (CDF)

```python
acc = [dist.cdf(valor) for valor in valores]
plt.plot(valores, acc, label='CDF', color='orange')
plt.title('Função Distribuição Acumulada - Exponencial')
plt.xlabel('Valor')
plt.ylabel('Probabilidade Acumulada')
plt.legend()
plt.grid(True)
plt.show()
```


### Exemplo: O tempo de espera em uma fila segue uma distribuição exponencial. Se o tempo médio esperado pelo cliente é de 10 minutos para ser atendido, qual a probabilidade:

### a) De que um cliente demore menos do que 12 minutos para ele ser atendido?
### b) De que um cliente demore entre 7 e 12 minutos para ser atendido?

```python

### Exemplo prático: tempo em fila (minutos) segue exponencial com média 10 minutos
lamb = 1/10  # taxa (lambda) é o inverso da média

# a) Probabilidade de o cliente demorar menos que 12 minutos (P(X <= 12))
prob_a = expon.cdf(12, loc=0, scale=1/lamb)
print(f'P(X <= 12 minutos): {prob_a:.4f}')
```

```python
# Cálculo manual para verificação
prob_a_manual = 1 - math.exp(-lamb*12)
print(f'P(X <= 12 minutos) manual: {prob_a_manual:.4f}')
```

```python

# b) Probabilidade de o cliente demorar entre 7 e 12 minutos
prob_b = expon.cdf(12, loc=0, scale=1/lamb) - expon.cdf(7, loc=0, scale=1/lamb)
print(f'P(7 <= X <= 12 minutos): {prob_b:.4f}')
```

```python
# Cálculo manual para verificação
prob_b_manual = math.exp(-lamb*7) - math.exp(-lamb*12)
print(f'P(7 <= X <= 12 minutos) manual: {prob_b_manual:.4f}')
```
